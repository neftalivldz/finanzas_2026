```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Configuración
ticker_symbol = 'AAPL'
start_date = '2022-01-01'
end_date = datetime.now().strftime('%Y-%m-%d')

print(f"Extrayendo datos de {ticker_symbol}")
print(f"Período: {start_date} hasta {end_date}")
```

```python
# Descargar datos históricos de Apple
apple_data = yf.download(ticker_symbol, start=start_date, end=end_date)

print(f"\nTotal de registros: {len(apple_data)}")
print("\nPrimeros 5 registros:")
print(apple_data.head())
```

```python
# Extraer precios de cierre
closing_prices = apple_data['Close']

print("\nÚltimos 10 precios de cierre:")
print(closing_prices.tail(10))
```

```python
# Visualización de precios de cierre
plt.figure(figsize=(16, 8))
plt.plot(closing_prices.index, closing_prices.values, linewidth=2, color='#007AFF', label='Precio de Cierre')

# Línea de precio promedio
plt.axhline(y=closing_prices.mean(), color='red', linestyle='--', linewidth=1.5, alpha=0.7, label=f'Promedio: ${closing_prices.mean():.2f}')

plt.title(f'Precio de Cierre de Apple (AAPL) - Enero 2022 a {end_date}', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Fecha', fontsize=14, fontweight='bold')
plt.ylabel('Precio de Cierre (USD)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(fontsize=12, loc='best')
plt.tight_layout()
plt.show()
```

```python
# Vista previa de los datos en formato DataFrame
print("\nResumen estadístico completo:")
print(closing_prices.describe())
```
