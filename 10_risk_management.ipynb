```python
from backtesting import PortfolioBacktester

bt = PortfolioBacktester(
        portfolio_path="portfolio.json",
        start_date="2025-12-22",
    )
pv = bt.run()
```

```python
df = bt.get_full_dataframe()
df
```

```python
import numpy as np
# Calcular los retornos diarios
portfolio_returns = df['portfolio_value'].pct_change().dropna()
sp500_returns = df['SP500'].pct_change().dropna()
# Alinear índices
portfolio_returns, sp500_returns = portfolio_returns.align(sp500_returns, join='inner')
# Beta = Cov(Rp, Rm) / Var(Rm)
cov_matrix = np.cov(portfolio_returns, sp500_returns)
beta = cov_matrix[0, 1] / cov_matrix[1, 1]
print(f"Beta del portafolio: {beta:.4f}")
```

```python
import lseg.data as ld
import pandas as pd
# Abrir sesión (requiere credenciales configuradas o Eikon Desktop abierto)
ld.open_session()
# Parámetros
ric = "1SPXFM6"
start_date = "2025-04-28"
end_date = "2026-04-27"
# Descargar serie de precios al cierre
df_f = ld.get_history(
    universe=ric,
    fields=["BID"],
    start=start_date,
    end=end_date,
    interval="1D"
)
df_f.index = pd.to_datetime(df_f.index)
df_f.columns = [ric]
df_f
```

```python
import numpy as np
# Retornos diarios de cada serie
futures_returns = df_f['1SPXFM6'].pct_change().dropna()
sp500_returns   = df['SP500'].pct_change().dropna()
# Alinear por índice de fecha (inner join)
futures_returns, sp500_returns = futures_returns.align(sp500_returns, join='inner')
# Beta = Cov(Rf, Rm) / Var(Rm)
beta = futures_returns.cov(sp500_returns) / sp500_returns.var()
print(f"Beta de 1SPXFM6 vs SP500: {beta:.4f}")
```
